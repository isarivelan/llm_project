```python

from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://jobs.nike.com/job/R-45775")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in JSON format
    containing the following keys: `role`, `expreience`, `skills`, `responsibilities.
    Only return the valid JSON and nothing else.
    """

)
chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data': page_data})
print(res.content)

```

```python
from langchain_core.output_parsers import JsonOutputParser

json = JsonOutputParser()
json_parser = json.parse(res.content)
json_parser
```

```python
import pandas as pd

df = pd.read_csv("my_portfolio.csv")
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient(path="db")
collection = client.get_or_create_collection(name="my_portfolio")

if not collection.count():
  for _, row in df.iterrows():
    collection.add(
        documents=row["Techstack"],
        metadatas={"links": row["Links"]},
        ids=[str(uuid.uuid4())]
    )
```

```python
job = json_parser

links = collection.query(
    query_texts=job["skills"],
    n_results=2).get('metadatas',[])
links
```

```python
prompt_email = PromptTemplate.from_template(
    """
    ### JOB DESCRIPTION:
    {job_description}
    ### INSTRUCTION:
    You are Isarivelan, a business development executive at AtliQ. AtliQ is an AI & Software Consulting company dedicated to facilitating
        the seamless integration of business processes through automated tools.
        Over our experience, we have empowered numerous enterprises with tailored solutions, fostering scalability,
        process optimization, cost reduction, and heightened overall efficiency.
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of AtliQ
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Atliq's portfolio: {link_list}
        Remember you are Isarivelan, BDE at AtliQ.
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):

        """
)
chain_email = prompt_email | llm
res = chain_email.invoke(input={'job_description': job, 'link_list': links})
print(res.content)
```
